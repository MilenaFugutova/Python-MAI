```python
three = 'Fizz'
five = 'Buzz'
tf = 'FizzBuzz'
for element in range (1, 101, 1) :
    if (element) % 3 == 0 and (element) % 5 == 0 :
        print (tf)
    elif (element) % 5 == 0 :
        print (five)
    elif (element) % 3 == 0 :
        print (three)
    else :
        print (element)
```
